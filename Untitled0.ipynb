```
#Modulos Necesarios
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
```

```
#Creacion de Dataset
n = 500 #Numero de Registros
p = 2 #Numero de Caracteristicas por cada Registros

X, Y = make_circles(n_samples=n, factor=0.5, noise= 0.05)

plt.scatter(X[Y == 0, 0], X[Y == 0, 1], c="black")
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], c="grey")
plt.axis("equal")
plt.show()
```
